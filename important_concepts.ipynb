```python
#decorators
""" adding a special feature to a already existing funcion , like wrapper a function with
another function"""

def order(func):
    def wrapper(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return wrapper
@order
def div(a,b):
  
    return a/b

def sub(a,b):
    return a-b

result1=div(2,4)
result2=order(sub)
print(result1)
print(result2(2,3))
```

```python
def decorator(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper
@decorator
def func():
    print("hai")

func()
```

# recursion
The ONLY 3 questions you must answer

Every recursion problem is solved by answering these:

1️⃣ What is the base case?

→ When should the function STOP?

2️⃣ How do I make the problem smaller?

→ Call the function again on smaller data.

3️⃣ How do I combine the results?

→ Add / count / return.

```python
nums = [10, 20, 30,4]
   
def copp(num):
    
    if not num:
        return 0
    num.pop()
    
 
    return 1+copp(num)
print(copp(nums))
```

```python
n=[1,2,3,4]
def sum1(n):

    if not n:
        return 0
    return n[0]+sum1(n[1:])
sum1(n)
```

```python
def keycount(d):
    if not d:
        return 0
    total=0
    for key,value in d.items():
        total+=1
        if isinstance(value,dict):
            total+=keycount(value)
    return total
        

data = {
    "A": {
        "B": {1:"b"},
        "C": {
            "D": {}
        }
    }
}
keycount(data)
```
