# Searching Algorithms


1> LinearSearch

```python
def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i]==target:
            return i
    else:
        return -1
l=[0,1,2,3,4,5,6,7]
target1=4
target2=9
print(f"index of {target1} = {linear_search(l,target1)}")
print(f"not found" if linear_search(l,target2)<0 else "founded")
```

2>Binary Search

```python
def binsearch(l,target):
    left=0
    right=len(l)-1

    while(left<=right):
        mid=(left+right)//2
        if l[mid]==target:
            return mid
        elif l[mid]<target:
            left=mid+1
        else:
            right=mid-1
    return -1
print(f"index of {target1} = {binsearch(l,target1)}")
print(f"{target2} not found" if binsearch(l,target2)<0 else "founded")

```

3> BinarySearch using recursion

```python

def bin_search(l,left,right,target):
    if left>right:
        return -1
    mid=(left+right)//2
    if l[mid]==target:
        return mid
    elif l[mid]<target:
        return bin_search(l,mid+1,right,target)
    else:
        return bin_search(l,left,mid-1,target)

print(f"index of {target1} = {bin_search(l,0,len(l)-1,target1)}")
print(f"{target2} not found" if bin_search(l,0,len(l)-1,target2)<0 else "founded")
```
