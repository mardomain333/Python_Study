# Sorting..

1>Bubble Sort


```python
def bubble_sort(l):

    n=len(l)
    for i in range(n):
        for j in range(0,n-i-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
l=[8,4,2,4,5,6]
bubble_sort(l)
print(l)
  
```

2>Selection Sort

```python

def Selection_sort(l):

    for i in range(len(l)-1):
        j=i+1
        key=j
        while(j<len(l)):
           if l[j]<l[key]:
               key=j
           j+=1
        l[i],l[key]=l[key],l[i]

l=[8,4,2,4,5,6]
Selection_sort(l)
print(l)              
               
 
```

3>Insertion Sort

```python
def Insertion_sort(l):
    for i in range(1,len(l)):
        last=i-1
        key=l[i]
        while last>=0 and l[last]>key:
            l[last+1]=l[last]
            last-=1
        l[last+1]=key
l=[8,4,2,4,5,6]
Insertion_sort(l)
print(l)
```
