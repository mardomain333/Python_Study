```python
arr=[5,2,2,1,1,1,1,4]
k=4
#here we are going to find the subarray with sum = k

prefix=0
current_index=-1
di={}
di[prefix]=current_index      #this will store the prefix:index 
                    #{0: -1, 5: 0, 7: 1, 9: 2, 10: 3, 11: 4, 12: 5, 13: 6, 17: 7}
maxcount=0
mincount=999
for a in arr:
    prefix+=a
    current_index+=1        
    dono=prefix-k           #dono mean  ippolate prefixsum il inn k mattiyal oru value 
                             #ath prefix dict ondel coco kandpidkam current-dict[dono]
    if dono in di.keys():
        coco=current_index-di[dono]
        maxcount=max(maxcount,coco)     #store current coco if greater else maxvalue remains same 
        mincount=min(mincount,coco)
        print(arr[di[dono]+1:current_index+1])
    di[prefix]=current_index   #adding current prefixsum:current_index to dict 
    
print(di)
print(f"max count={maxcount}")
print(f"min count={mincount}")


```

3> fixed size sliding window
   find the maximum sum of subarray k

   

```python
arr = [1,2,3,1,2,-3,2,3,3]
k = 3

window_sum = sum(arr[:k])
max_sum = window_sum
l = 0
maxs=arr[l:k]
for r in range(k, len(arr)):
    window_sum += arr[r]
    window_sum -= arr[l]
  
    l += 1
    if window_sum>max_sum:
        maxs=arr[l:r+1]
    max_sum = max(max_sum, window_sum)

print(max_sum)
print(maxs)
```

4> longest substring without repeating characters 
 

```python
s = "abcabcbb"
longest=0
set_repeated=set()
l=0
for r in range(len(s)):
    
    while(s[r] in set_repeated ):
        set_repeated.remove(s[l])
        l+=1
    set_repeated.add(s[r])
    current_len=(r-l)+1
    longest=max(longest,current_len)
print(longest)
```

```python
arr = [5, -2, 3, 4, -1]

max_sum = arr[0]
current_sum = arr[0]

start = end = temp_start = 0

for i in range(1, len(arr)):
    if arr[i] > current_sum + arr[i]:
        current_sum = arr[i]
        temp_start = i
    else:
        current_sum += arr[i]

    if current_sum > max_sum:
        max_sum = current_sum
        start = temp_start
        end = i

print("Max Sum:", max_sum)
print("Subarray:", arr[start:end+1])
print("Indices:", start, end)
```

6> Two sum


```python
def twoSum(nums, target):
    seen = {}

    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            return [seen[diff], i]
        seen[num] = i
nums = [2, 7, 11, 15]
target = 9
a,b=twoSum(nums,target)
print(nums[a:b+1])
```
