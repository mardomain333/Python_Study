# list comprehension

```python
lst=[]
for i in range(11):
    lst.append(i)
print(lst)
#this same thing can be done by list comprehesion easily
newlist=[n for n in range(11)]
print(newlist)
if newlist==lst:
    print("look both are same")
```

```python
#we can perform expression also 
lst=[1,2,3,4,5]
newlist=[x**2 for x in lst]
print(newlist)
oddlist=[x for x in lst if x&1]
print(oddlist)

names=['gopal','ramu','ratha','pathuma']
l=[name.capitalize() for name in names if name[0]=='r']
print(l)
```

```python
l=[]
for i in range(2):
    for j in range(2):
        l.append((i,j))
print(l)
#this can also depicted using list comprehension

newlist=[(i,j) for i in range(2)
                for j in range(2)]
print(newlist)
#fantastic alle 

#not only list we have set comprehension also
s={x for x in range(10)}
print(s)

#we can also do dictionary also
d={a:a**2 for a in s}
print(d.items())

#but if we try to do with tuple its generate a generator object
t=(a for a in range(10))
print(t)
```

Generator
A generator is a function that returns an iterator using yield, producing values lazily one at a time.
A generator is a special function that:

Produces values one at a time

Uses yield instead of return

Remembers its state between calls

ðŸ‘‰ It is memory-efficient.

we can do it :
 

```python
def gen(n):
    for a in range(n):
        return a
    
print(gen(4))
# we got only 0 why because return will stop the function execution 
#here we can use generator using yeild keyword
def gen(n):
    for a in range(n):
        yield a
gen=gen(5)
# ippo iterator anel full number um orumich set akki vekkum 
# but generator angane alla vendunne anusarich therum 
print(next(gen)) #0
print(next(gen)) #1
print(next(gen)) #2

#now we can use t in previous 
for a in t:
    print(a)
```

```python
#chaining generator
password=['hai','hello','fastest']
def length(itr):
    for a in itr:
        yield a
def adder(itr):
    for a in itr:
         yield '8'*len(a)
for a in adder(length(password)):
    print(a)
```

```python
#Enumerator 
l=["milk","rice","soap","shirt"]
for i in range(len(l)):
    print(f"{i+1} {l[i]}")
# this thing can be done using enumerator fn
for i,item in enumerate(l,start=1):
    print(i,item)
```

now we can talk about zip and zip_longest

```python
from itertools import zip_longest
name=['gokul','akul','ananthu','lekshmi','revu','reshma']
roll=[99234,2213,3422,4234,5242]
#if i want to iterate throug both list we can use index right

for i in range(min(len(name),len(roll))):
    print(name[i],roll[i])
#look at here it printed upto revu only because upto revu only value in roll
#this thing can be done using fun called zip
for i,j in zip(name,roll):
    print(i,j)

#here also only min len is considered if we want to do with longest thing
#we can use zip_longest() from itertools
for i in zip_longest(name,roll,fillvalue=0):
    print(i)
    #reshma got none value we can set any value insteat of none using fillvalue


```

```python
#cycle 
from itertools import cycle
#cycle an iterable 
l=[1,2,3,4,5]
l="gokul"
count=0
for i in cycle(l):
    count+=1
    print(i)
    if count==10:
         break
```

```python
from itertools import repeat
#repeat
# same thing like cycle but not reapeating the elements but the list itself
s=[1,2,3]
for i in repeat(s,times=3):
    print(s)                #actually i dont like it 

for i in range(len(s)):
    print(s)
```

```python
#combinations 
from itertools import combinations
 #we get different not repeatable combination 
l=[1,2,3,4,5]
l="gokul"
for i in combinations(l,4):
    print(i)
```

```python
#permutationss
from itertools import permutations
l=[1,2,3,4,5]
l="gokul"
for i in permutations(l,2):
    print(i)
```

```python
#product
from itertools import  product
#assume the concept of cross product 
a=[1,2,3,4]
b=[5,6]            # a*b
count=0
for i in product(a,b) :
    print(i)
    count+=1
print(count)

#use 
a="gokul"
b="akul"
for i in product(a,b):
    print(i[0]+i[1])
```

```python
#chain

from itertools import chain
l=[1,2,3,4,5]
l1=(c for c in range(10))
names=["gokul","akul","rahul"]

# chain can be used to combine these things
# l+l1+names actully l1 is a generator

for i in chain(l,l1,names):
    print(i)
```

```python
from itertools import filterfalse
#map 
l=[1,2,3,4,5]
new=list(map(lambda x:x*x,l))
print(new)
#filter 
new=list(filter(lambda a:a&1,l))
print(new)
#filterfalse
new=list(filterfalse(lambda a:a&1,l))
print(new)

#reduce
from functools import reduce
new=reduce(lambda x,y:x+y,l)
print(new)

#accumulates
from itertools import accumulate

for i in accumulate(l,lambda a,b:a+b):
    print(i)                              #cumulative things
```

```python
#compress
from itertools import compress
name=["gokul","akul","rahul","annathu"]
is_adult=[True,False,True,True]
for i in compress(name,is_adult):
    print(i)
```
